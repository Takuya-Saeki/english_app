```python
import csv

# CSVファイルのパス
file_path = '../word_list/csv_folder/word_list_2zi_test.csv'
```

```python
# 単語を格納するリストを初期化
words = []

# CSVファイルを開く
with open(file_path, newline='') as file:
    reader = csv.reader(file)

    # 各行を読み込み、単語をリストに追加
    for row in reader:
        # rowはリスト形式なので、最初の要素を取得
        word = row[0]
        words.append(word)

# 結果を表示
print(words)
```
